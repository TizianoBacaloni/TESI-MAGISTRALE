## NON SI PUò CREARE UNA LOOP NELLA QUALE SI CALCOLANO MEDIA E STANDARD DEVIATION PER I DIVERSI PAESI MENTRE SI CERCA CONTEMPORANEAMENTE DI MODIFICARE IL DATABASE, SERVONO DUE LOOP SEPARATE MA UGUALI COME INTESTAZIONE:



***DI FATTO CREO UN DIZIONARIO VERO E PROPRIO CON DUE LIVELLI DI CHIAVE:***

**1) [PAESE] MI PERMETTE DI INSERIRE NEL DIZIONARIO IN POSIZIONE [PAESE] TUTTE LE STATISTICHE RELATIVE**

**2) SELEZIONATA LA POSIZIONE CHE VOGLIO, IL [PAESE] CHE MI SERVE, SCELGO LA STATISTICA COME 'STATISTICA**

```python
group_data = {}
for paese in sample:

    media_price = house.loc[str(paese),'Real house price indices'].mean()
    media_pti = house['Price to income ratio'].loc[str(paese)].mean()
    
    std_price = house['Real house price indices'].loc[str(paese)].std()
    std_pti = house['Price to income ratio'].loc[str(paese)].std()

    group_data[paese]={

        'media_price': media_price,
        'media_pti': media_pti,
        
        'std_price': std_price,
        'std_pti': std_pti
    }

group_data['Belgium']
```

**PER OGNI PAESE ESTRAGGO LE INFORMAZIONI E LE USO PER MODFICIARE IL DATABASE**

```python
standardized_data = []
standardized_data_pti = []
for paese in sample:

    media_price = group_data[paese]['media_price']
    std_price = group_data[paese]['std_price']
    media_pti = group_data[paese]['media_pti']
    std_pti = group_data[paese]['std_pti']

    dati = (house.loc[str(paese)]['Real house price indices'] - media_price)/std_price
    dati_pti = (house.loc[str(paese)]['Price to income ratio']- media_pti) /std_pti
    
    standardized_data.append(dati)
    standardized_data_pti.append(dati_pti)
    
    group = pd.concat(standardized_data)
    group_pti = pd.concat(standardized_data_pti)


group.reset_index(inplace=True)
group_pti.reset_index(inplace=True)
house.reset_index(inplace=True)
```

```python

house['Real house price index']= group['Real house price indices']
house ['Price-Income Ratio']  = group_pti['Price to income ratio']


house= house.drop(['Real house price indices','Price to income ratio'],axis=1)
```
