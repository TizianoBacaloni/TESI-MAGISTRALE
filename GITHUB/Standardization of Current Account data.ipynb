**CONSIDERO SOLO LE COLONNE 'COUNTRY','TIME','VALUE', NON TENENDO CONTO DELLA FREQUENZA DATO CHE I VALORI SONO GIà TRIMESTRALI**

```python
ca = ca[['Country','Time','Value']]
ca.head(10)
```

```python
nazioni = ca.groupby('Country').sum().index
nazioni

anni = ca.groupby('Time').sum().index
anni[0]
```

```python
ca.set_index(['Country','Time'],inplace=True)
```

**CONTROLLO DELLE PRIME ED ULTIME RILEVAZIONI PER CIASCUN PAESE PRESENTE ALL'INTERNO DEL CAMPIONE**



***DANIMARCA,RUSSIA,GRECIA,NETHERLANDS,POLONIA ED IRLANDA HANNO PRIME RILEVAZOINI DIVERSE DAGLI ALTRI PAESI***

```python
ca.loc[('Australia','Q1-2000')]

for nazione in nazioni:
    
    naz = ca.loc[nazione]
    anni = list(naz.groupby('Time').sum().index)
    primo = anni[0]
    ultimo = anni[-1]
    print(f' per {nazione} il primo anno è {primo} ed ultimo è {ultimo}')
    
        
        
```

**ELIMINAZIONE DI DENMARK E RUSSIA DALLA LISTA DELLE NAZIONI**


```python
nazioni = nazioni.drop(['Denmark','Russia','Greece','Netherlands','Ireland','Poland'])
integrande = ['Denmark','Russia','Greece','Netherlands','Ireland','Poland']
```

**PER CIASCUNA NAZIONE SI CALCOLANO MEDIA E STD  PER OGNI PAESE E LE SI INSERISCE IN UN APPOSITO DATABASE IN POSIZIONE 'PAESE'**

```python
data = {}



for nazione in nazioni:
     
    mean = ca.loc[nazione].mean()
    std = ca.loc[nazione].std()

    data[nazione] ={

        'mean': mean,
        
        'std': std
        }


for nazione in integrande:
    mean = ca.loc[nazione].mean()
    std = ca.loc[nazione].std()

    data[nazione] ={

        'mean': mean,
        
        'std': std
        }

data
```

**STANDARDIZZAZIONE DATI PER TUTTE LE NAZIONI,DISTINGUENDO I CICLI A SECONDA DELLA DATA DI INIZIO DELLE RILEVAZIONI**

```python
for nazione in nazioni:
    for anno in anni:
     
        
        ca.loc[(nazione,anno)] = (ca.loc[(nazione,anno)] - data[nazione]['mean'])/ data[nazione]['std']
        
for nazione in  integrande:

    anni_int= list(ca.loc[nazione].index)
    primo = anni[0]
    ultimo = anni[-1]
    print(f'per la {nazione} il primo anno coinicide con {primo} e lo ultimo con {ultimo}')

    for anno in anni_int:

        ca.loc[(nazione,anno)] = (ca.loc[(nazione,anno)]-data[nazione]['mean'])/data[nazione]['std']
    
        
   


```

```python
can_bel = pd.read_csv('/Users/tiziana/Desktop/PYTHON/TESI MAGISTRALE/DATI/CA BALANCE/BEL E CAN OECD CURRENT ACCOUNT.csv')
can_bel.head()
```

```python
can_bel = can_bel.set_index('Country')
can = can_bel.loc['Canada']
can = can[['TIME','Value']]

can = can.reset_index()
can.set_index(['Country','TIME'],inplace=True)

mean = can.mean()
std = can.std()

can = (can - mean) / std
```

```python

bel = can_bel.loc['Belgium']
bel = bel[['TIME','Value']]

bel = bel.reset_index()
bel.set_index(['Country','TIME'],inplace=True)

mean = bel.mean()
std = bel.std()

bel = (bel - mean) / std
```

**STESSA PROCEDURA MA PER LE NAZIONI CON PRIMA OSSERVAZIONE IN PERIODI DIVERSI**

```python
ca = pd.concat([ca,bel,can],axis=0)
```
