## SCELGO IL NUMERO DI COMPONENTI PRINCIPALI BASANDOMI SULLA VARIANZA CUMULATA SPIEGATA COMPONENTI

```python
from sklearn.decomposition import PCA

Pca_model = PCA()
Pca_model.fit(financial)


cumulative_variance = np.cumsum(Pca_model.explained_variance_ratio_)
n_components = np.argmax(cumulative_variance >= 0.80) + 1
```

```python
cumulative_variance
```

```python
explained_variance_ratio = Pca_model.explained_variance_ratio_
print(explained_variance_ratio)
```

```python
print(list(reversed(explained_variance_ratio)))
```

```python
plt.plot(np.arange(1,8,1),explained_variance_ratio,marker='o')
plt.xlabel('Number of Components')
plt.ylabel('Variance')
plt.title('Scree Plot and Single Variance')
```

```python
plt.plot(np.arange(1, len(cumulative_variance) + 1), cumulative_variance, marker='o', linestyle='-', color='r', label='Cumulative Variance')
plt.xlabel('Number of Components')
plt.ylabel('Variance')
plt.title('Scree Plot and Cumulative Variance')
```

## CLUSTER ANALYSIS NEL 2007

```python
data = dataset.set_index(['Trimestre'])
data07 = data.loc[['2007Q1', '2007Q2', '2007Q3', '2007Q4']]
data07.reset_index(inplace=True)

data07.set_index(['Country','Trimestre'],inplace=True)
```

```python
database = []
for nazione in data07.groupby('Country').sum().index:
    dati_nazionali = data07.loc[nazione]
    dati_nazionali['Country'] = nazione
    database.append(dati_nazionali)
    dataset07 = pd.concat(database,axis=0)

dataset07.reset_index(inplace=True)
dataset07.set_index(['Country','Trimestre'],inplace=True)
```

```python
dataset07 = dataset07.groupby('Country').mean()
```

```python

model = AgglomerativeClustering(n_clusters=None,distance_threshold=0.3)
cluster_labels = model.fit_predict(dataset07)
```

```python
from scipy.spatial.distance import pdist
dist_matrix = pdist(dataset07, metric='euclidean')

linkage_matrix = hierarchy.linkage(dist_matrix, method='ward')
```

```python
plt.figure(figsize=(19, 15))
hierarchy.dendrogram(linkage_matrix,labels = dataset07.index)
```

```python
distanze = linkage_matrix[:, 2]
plt.figure(figsize=(10, 5))
plt.plot(range(1, len(distanze) + 1), sorted(distanze, reverse=True))
```

## GUARDANDO IL GRAFICO PARE CONFERMATO CHE , UTILIZZANDO LA METRICA EUCLIDEA PER IL CALCOLO DELLE DISTANZE, E FACENDO RIFERIMENTO AL METOODO 'SINGLE' I GRUPPI DEBBANO ESSERE 77

```python
cutoff=7
from scipy.cluster.hierarchy import fcluster
clusters = fcluster(linkage_matrix,cutoff,criterion = 'maxclust')

clusters_df = pd.DataFrame({'Paese': dataset07.index, 'Gruppo': clusters})
print(clusters_df.sort_values('Gruppo'))
```
