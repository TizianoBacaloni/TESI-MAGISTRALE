## CAMBIO NOME AD UNA SPECIFICA COLONNA DEL DATABASE E CONCATENO

**LO FACCIO CON DATABASE.RENAME SPECIFICANDO CHE VOGLIO AGIRE SULLE (COLONNE) ED IN PARTICOLARE FARE SI CHE {LA COLONNA CHE PRIMA SI CHIAMA X : DOPO SI CHIAMI Y}**



```python

for nazione in nazioni_domestic_credit:
     nazione.rename(columns={nazione.columns[1] : 'Credit to private NF sector'},inplace = True)


prv_cre = pd.concat(nazioni_domestic_credit,axis=0)

dict = {'NaN' : np.nan}
prv_cre['Credit to private NF sector'].map(dict)
na = prv_cre['Credit to private NF sector'].isna().sum()
na
```

```python
prv_cre['DATE'] = pd.to_datetime(prv_cre['DATE'])
prv_cre['Trimestre']= prv_cre['DATE'].dt.to_period('Q')
```

```python
prv_cre = prv_cre[['Country','Trimestre','Credit to private NF sector']]
```

```python
nazioni = list(prv_cre.groupby('Country').sum().index)
anni = list(prv_cre.groupby('Trimestre').sum().index)

prv_cre.set_index(['Country','Trimestre'],inplace=True)

stat = {}
for nazione in nazioni:
    mean = prv_cre.loc[nazione].mean()
    std = prv_cre.loc[nazione].std()

    stat[nazione] = {
        'mean': mean,
        'std': std
    }
```

```python
prv_cre.loc['Netherlands']
```

```python
for nazione in nazioni:
    for anno in anni:
        prv_cre.loc[(nazione,anno)] = (prv_cre.loc[(nazione,anno)] - stat[nazione]['mean'] ) / stat[nazione]['std']
```

```python
prv_cre.to_csv('domestic credit.csv')
```
