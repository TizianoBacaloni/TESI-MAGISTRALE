**IMPORTANTE COME IN QUESTO CASO PER POTER TRASFORMARE LA COLONNA DELLE APERTURE IN NUMERICA COSì DA POTERNE CALCOLARE LA MEDIA SIA NECESSARIO SPECIFICARE str.replace() MENTRE AD ESEMPIO NEL CASO DELL'AUSTRIA NON SERVE**

```python
Australia = pd.read_csv('/Users/tiziana/Desktop/PYTHON/TESI MAGISTRALE/DATI/FINANCIAL INDEX/STOCK MARKET /MARKET INDEXES/AUSTRALIA.csv')
Australia.head()
Australia['Date'] = pd.to_datetime(Australia['Date'])
Australia['trimestre'] = Australia['Date'].dt.to_period('Q')

Australia['Open'] = Australia['Price'].str.replace(',', '').astype(float)
pd.to_numeric(Australia['Open'])

#SERVE A CALCOLARE IL VALORE MEDIO PER I DIVERSI GRUPPI DA 3 OSSERVAZIONI MENSILI, COSì DA OTTENERE I VALORI TRIMESTRALI
mean_price = Australia.groupby('trimestre')['Open'].mean()
Australia.head()

Aus = pd.DataFrame(mean_price)
Aus =Aus.reset_index()

Aus.columns = ['DATE', 'Stock idx open-price']
mean = Aus['Stock idx open-price'].mean()
std = Aus['Stock idx open-price'].std()

Aus['Stock idx open-price'] = (Aus['Stock idx open-price'] - mean) /std
Aus['Country'] =list(['Australia'] * Aus.shape[0])
```
